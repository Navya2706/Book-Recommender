```python
##Poplarity Based Recommender System
```

```python
#Merging books and ratings
ratings_with_books=ratings.merge(books, on='ISBN')
```

```python
ratings_with_books
```

```python
num_rating_df = ratings_with_books.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df
```

```python
num_rating_df.rename(columns={'Book-Rating':'num_rating'}, inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_books.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df
```

```python
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'}, inplace=True)
avg_rating_df
```

```python
popularity_df= num_rating_df.merge(avg_rating_df, on='Book-Title')
popularity_df
```

```python
popularity_df =popularity_df[popularity_df['num_rating']>250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular.shape
```

```python
popular= popularity_df.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_rating','avg_rating']]
popular
```

```python
#COLLABORATIVE FILTERING
```

```python
x = ratings_with_books.groupby('User-ID').count()['Book-Rating']>200
users=x[x].index
```

```python
filtered_rating = ratings_with_books[ratings_with_books['User-ID'].isin(users)]
filtered_rating
```

```python
popular['Image-URL-M'][0]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books= y[y].index
```

```python
famous_books
```

```python
final_rating=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_rating
```

```python
pt=final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score= cosine_similarity(pt)
similarity_score[0]
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    #index fetch
    index = np.where(pt.index==book_name)[0][0]
    distances= similarity_score[index]
    similar_items= sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1], reverse=True)[1:6]
    
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend('Message in a Bottle')
```
